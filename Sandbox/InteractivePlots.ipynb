```julia
using Plots
plotlyjs()
using PlotlyJS: relayout
```

```julia
xd = rand(-400:0.01:400.0,45)
yd = rand(0:0.01:100.0, length(xd))

x = range(-400, stop = 400.0, length = 10000)
y = sum(yd[i] .* exp.(-(x .- xd[i]).^2 / 1) for i in 1:length(xd))
text = ["|J'=$(5i),ℓ'=$(i),Σ'=$(-i/2)> ← |J''=$(4i),ℓ''=$(-i),Σ''=$(i/2)>" for i in 1:length(xd)]

p = plot(xd, yd, linealpha=0, hover = text, frame=:box, size=(1000,500))
plot!(p,x,y, label="Spectrum", hover = false, legend=false)
xlabel!("Frequency")
ylabel!("Intensity")
PlotlyJS.relayout(Plots.plotlyjs_syncplot(p), hovermode="x", hoverdistance=5)
```
