Code to test bending mode implementation.

```julia
using Revise
```

```julia
using JuliaSpectra
using Plots
```

```julia
bounds = (l=[-1,1], J = 1/2:7/2, I=1/2, Σ=[-1/2,1/2], Λ=0, S = 1/2, F=0:4, M=-4:4)
basisX = enumerate_states(LinearCaseA_Bend_Field, bounds)

bounds = (l=[-1,1], J = 1/2:5/2, I=1/2, N=1:2, Λ=0, S = 1/2, F=0:4, M=-4:4)
basisX = enumerate_states(LinearCaseB_Bend_Field, bounds)
;
```

```julia
const sol = 299792458

const B = 7348.40052
const γ = -81.1538184
const qv = -20.0
const b=4.80
const c = 2.46
const μB = 1.4
const gS = 2.0023
const μel = 1.9 * 0.5034
;
```

```julia
Elist = range(1e-8, stop=250, length=1000)

# X State
Hrot = B * Rotation
Hsr = γ* SpinRotation
Hhf = (b+c/3) * HyperfineFermi + c * Hyperfine_Dipolar_c
Hld = qv * lDoubling
HX0 = Hrot + Hsr + Hhf + Hld
eigsysX = eigenshuffle(HX0, μel*Stark, basisX, Elist, pack_eigensystem=true)
;
```

```julia
@time eigenshuffle(HX0, μel*Stark, basisX, Elist, pack_eigensystem=true)
```

```julia
plotZeemanlevels(eigsysX[1:24,end], energy_offset = 7355, units="cm-1", ylabel="Energy", legend=false, lw=2, color=:black)
```

```julia
plot(Elist,eigsysX[1:24,:].E',legend=false, lw=2)
```
